```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
```

```python
x=np.linspace(-10,10,100)
y=math.log(2) + 0.5 * abs(x-1)-abs(x)
```

```python
fig = plt.figure(figsize = (10,10));
plt.plot(x,y,label = 'class conditional PDF')
plt.title('log-likelihood ratio Graph',fontsize = 15)
plt.xlabel('Random Variable X',fontsize = 15)
plt.ylabel('l(x)',fontsize = 15);
#plt.legend()
plt.legend(fontsize = 15)
plt.show()
```
