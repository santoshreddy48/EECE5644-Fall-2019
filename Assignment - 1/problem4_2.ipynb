```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.stats as stats
import math
```

```python
mu1 = 0
variance1 = 1
sigma1 = math.sqrt(variance1)
x1 = np.linspace(mu1 - 5*sigma1, mu1 + 5*sigma1, 100)

mu2 = 1
variance2 = 2
sigma2 = math.sqrt(variance2)
x2 = np.linspace(mu2 - 5*sigma2, mu2 + 5*sigma2, 100)
```

```python
fig = plt.figure(figsize = (10,10));
plt.title('Gaussian Distribution of prior',fontsize = 15)
plt.plot(x1, stats.norm.pdf(x1, mu1, sigma1),label = 'class 1 prior')
plt.plot(x2, stats.norm.pdf(x2, mu2, sigma2), label = 'class 2 prior')
plt.xlabel('x',fontsize = 15)
plt.ylabel('P(x|w)',fontsize = 15);
plt.axvline(x=0.839, ymax = 0.68,color = 'black', linestyle = 'dashed',label = 'decision boundry')
plt.axvline(x=-2.839,ymax = 0.066,color = 'black',linestyle = 'dashed')
plt.legend(fontsize = 15)
plt.show()
```

```python
x1 = np.linspace(mu1 - 5*sigma1, mu1 + 5*sigma1, 100)
p1 = stats.norm.pdf(x1, mu1, sigma1)
p2 = stats.norm.pdf(x1, mu2, sigma2)
fig = plt.figure(figsize = (10,10));
plt.title('normal distribution curves of posterior',fontsize = 15)
plt.plot(x1, p1/((p1+p2)), label = 'class 1 Posterior')
plt.plot(x1, p2/((p1+p2)), label = 'class 2 Posterior')
plt.xlabel('x',fontsize = 15)
plt.ylabel('P(l|w)',fontsize = 15);

plt.legend(fontsize = 15)
plt.show()
```
